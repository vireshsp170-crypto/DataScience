##### Exercise 4 : Problem Statement (Question): Hospital Billing System
##### A hospital charges its patients based on the following:
- Room charges per day: ₹2,000
- Doctor consultation fee (flat): ₹1,500
- Lab test charges: Based on the number of tests taken. Each test costs ₹300
- Medicine charges: Total cost of medicines provided
- Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied
  
Write a Python program to calculate the final bill for a patient who:
- Stayed for 4 days


###### S1: Problem Statement - Create a Python script that prints the Hospital Bill.
######    Requirements: To solve the given problem we should know how to multiply charges with the frequency(no. of stays/visits). and add all the various charges.
###### S2: Approach 
           We make use of operators in the python.
###### S3: Pseudocode - 
        Initialize constants
        Input patient details (given in problem)
        Calculate individual charges
        Calculate total bill before discount
        Check if discount is applicable
        Final bill after discount
        Display the result
###### S4: Code

```python
# Constants 
room_cost_per_day = 2000
consultation_fee = 1500
lab_test_cost = 300
discount_limit = 10000 
discount_percentage = 0.10

# Patient Deatils given in the problem statement
days_admitted = 4
tests_done = 3
medicine_cost = 2400

# Calculate Charges

total_room_cost = room_cost_per_day * days_admitted
total_consultation_fees = consultation_fee
total_lab_cost = lab_test_cost * tests_done

# Calculate bill before availing discount
total_bill = total_consultation_fees + total_lab_cost + total_room_cost + medicine_cost

# Give Discount if applicable
if total_bill > discount_limit:
    discount_amount =  total_bill*discount_percentage
else:
    discount_amount = 0

final_bill = total_bill - discount_amount

# Display Detailed Bill
print("----- Hospital Bill -----")
print(f"{'Room Charges:':25} {total_room_cost:>8}")
print(f"{'Consultation Fee:':25} {total_consultation_fees:>8}")
print(f"{'Lab Charges:':25} {total_lab_cost:>8}")
print(f"{'Medicine Charges:':25} {medicine_cost:>8}")
print("-" * 40)
print(f"{'Total Before Discount:':25} {total_bill:>8}")
print(f"{'Discount:':25} {discount_amount:>8.2f}")
print(f"{'Final Bill:':25} {final_bill:>8.2f}")
```
