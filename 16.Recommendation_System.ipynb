```python
# Checking unique values in the 'episodes' column
# to verify whether it behaves like a categorical feature

data['episodes'].unique()
```

```python
'''
The column contains numeric values stored as strings, and the presence of the value 'Unknown' results in the feature being interpreted as a categorical
variable.
'''
```

```python
# Replacing 'Unknown' values with NaN to handle them as missing values

data['episodes'] = data['episodes'].replace('Unknown', np.nan)
```

```python
# Converting the column to numeric datatype
data['episodes'] = pd.to_numeric(data['episodes'])
```

```python
# Checking the data type of the 'episodes' column
data['episodes'].dtype
```

```python
# Checking the number of missing values in each column
data.isna().sum()
```

```python
# Imputing missing values in 'episodes' using the median
data['episodes'] = data['episodes'].fillna(data['episodes'].median())

# Converting the column to integer type after imputation
data['episodes'] = data['episodes'].astype('int64')
```

```python
# Filling missing values in the 'genre' column with 'Unknown'
data['genre'] = data['genre'].fillna('Unknown')
```

```python
# Filling missing values in the 'type' column with 'Unknown'
data['type'] = data['type'].fillna('Unknown')
```

```python
# Filling missing values in the 'rating' column using the median value
data['rating'] = data['rating'].fillna(data['rating'].median())
```

```python
# Feature selection for recommendation system
# Converting 'genre' text data into numerical features using TF-IDF
# TF-IDF is more suitable than One-Hot Encoding for text data with multiple categories

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')
genre_tfidf = tfidf.fit_transform(data['genre'])
```

```python
# Normalizing numerical features ('episodes', 'rating', 'members')
# Min-Max scaling brings all values into the range [0, 1]

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
num_features = scaler.fit_transform(data[['episodes', 'rating', 'members']])
```

```python
# Combining TF-IDF genre features with normalized numerical features
# hstack is used because TF-IDF output is a sparse matrix

from scipy.sparse import hstack

final_features = hstack([genre_tfidf, num_features])
```

```python
# Computing cosine similarity between all items
# This measures how similar each item is to every other item

from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(final_features)
cosine_sim
```

```python
# Creating a title-to-index mapping (do this once, outside the function)
title_index = pd.Series(data.index, index=data['name'].str.lower()).drop_duplicates()

# Improved recommendation function
def recommend_anime(anime_title, dataset, similarity_matrix,
                    top_n=10, min_similarity=0.3):
    
    anime_title = anime_title.lower()
    
    # Check if anime exists
    if anime_title not in title_index:
        return "Anime not found in dataset."
    
    # Get index of selected anime
    anime_idx = title_index[anime_title]
    
    # Get similarity scores
    sim_scores = list(enumerate(similarity_matrix[anime_idx]))
    
    # Sort by similarity (highest first)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Remove itself (first entry) and apply similarity threshold
    sim_scores = [
        (idx, score) for idx, score in sim_scores[1:]
        if score >= min_similarity
    ]
    
    # Select top recommendations
    top_indices = [idx for idx, _ in sim_scores[:top_n]]
    
    # Return recommended anime details
    return dataset.loc[top_indices, ['name', 'genre', 'rating', 'episodes']]
```

```python
# Checking recommendations with a different similarity threshold
recommend_anime("Naruto", data, cosine_sim, top_n=5, min_similarity=0.3)
```

- #### User-Based Collaborative Filtering:
       Recommends items by identifying users with similar preferences and suggesting items that those similar users have liked.

- #### Item-Based Collaborative Filtering:
       Recommends items by finding items that are similar to those the user has previously liked or interacted with.

- #### Collaborative Filtering (General Concept):
       A recommendation technique that predicts a user’s interests based on the behavior and interactions of many users, 
       rather than relying on item content or features.
