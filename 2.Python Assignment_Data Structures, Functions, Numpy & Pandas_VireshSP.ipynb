#### 1.	Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
# Creating a list of 5 integers
integer_list = [3, 5, 7, 9, 2]

#Taking a copy of the original list
integer_list_cp = integer_list

print("Integer List: ", integer_list)
print('-'*50)

# append() : which adds an element to the list.
integer_list.append(4)
print('After Appending an element: ', integer_list)
print('-'*50)

# extend() : can add multiple values elements to the list
integer_list.extend([10,11])
print('After Extending elements: ',integer_list)
print('-'*50)

#insert() : Can insert/add elements at particular index
integer_list.insert(3, 0)
print('After Inserting an element: ',integer_list)
print('-'*50)

#remove() : Can remove elements
integer_list.remove(2)
print('After Removing an element: ',integer_list)
print('-'*50)

#pop() : Can delete element
integer_list.pop()
print('After popping an element: ',integer_list)
print('-'*50)

#pop(index) : Can deleting an  element from a aprticular index
element1 = integer_list.pop(2)
print('Element popped out: ',element1)
print('After popping an element at index: ',integer_list)
print('-'*50)


#index(element) : It Return first index of value/element
index_pos = integer_list.index(9)
print('Index position ofthe given element: ',index_pos)
print('-'*50)

#count() :  Return number of occurrences of value.
element_count = integer_list.count(3)
print('Occurance of the given element: ', element_count)
print('-'*50)

#sort(reverse=False) : Sort will arrange the elemnts inside the list in ascending order by default
integer_list.sort()
print('Sorted List in Ascending Order : ',integer_list)
print('-'*50)

#sort() : Sort will arrange the elemnts inside the list in ascending order by default
integer_list_cp.sort(reverse=True)
print('Sorted List in Descending Order : ',integer_list_cp)
print('-'*50)

#remove() : Can remove elements
integer_list.reverse()
print('After Reversing all the elements in the list : ',integer_list)
print('-'*50)

#clear() : Will remove all elements
integer_list.clear()
print('After Removing all elements from the List: ',integer_list)
print('-'*50)
```

#### 7.	Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.

```python
#Taking number input from the user in the num variable.
num = int(input('Enter a number: '))

#Checking for positive and even number.
if num>0:
    if num %2 == 0:
        print('Positive Even Number')
    else:
        print('Positive Number')
#Checking for zero or negative number
elif num == 0:
        print('Zero')
else:
        print('Negative')
```

#### 11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows


```python
import pandas as pd
```

```python
# Creating a DataFrame with 10 rows using dictionary.
edata = {
    'Name': ['Ram', 'Shyam', 'Charan', 'Dev', 'isha', 
             'Amit', 'Vinay', 'Ajay', 'Vijay', 'Amit'],  
    'Age': [25, 30, 35, 40, 28, 33, 29, 31, 36, 25],
    'City': ['New York', 'London', 'Paris', 'Berlin', 'Tokyo',
             'Sydney', 'London', 'Paris', 'Berlin', 'New York'],
    'Salary': [50000, 60000, np.nan, 80000, 75000,
               62000, 58000, np.nan, 70000, 50000]
}
#Converting array into dataframe using pandas module and dataframe().
e_df =pd.DataFrame(edata)
print(e_df)
```

```python
# Using .info(), .describe()
print('-------DataFrame Info-------')
e_df.info()
print('\n-----DataFrame Describe-----')
e_df.describe()
```

```python
# Selecting 'Name' and 'City' columns
print('\n-----Select Name and City-----')
e_df.loc[: ,['Name', 'City']]
```

```python
# Dropping 'City' column
df_dropped = e_df.drop(columns='City')
print('-----After Dropping City Column-----')
print(df_dropped)
```

```python
# Filling  missing values in 'Salary' column with the mean
e_df['Salary'].fillna(e_df['Salary'].mean())
print("\n----- After Filling Missing Salary -----")
print(e_df)
```

```python
# Removing duplicate rows
e_df_no_duplicates = e_df.drop_duplicates()
print("\n----- After Removing Duplicates -----")
print(e_df_no_duplicates)
```
