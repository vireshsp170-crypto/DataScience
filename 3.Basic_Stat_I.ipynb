### BASIC STATISTICS

#### Descriptive Analytics for Numerical Columns:
- Objective: To compute and analyze basic statistical measures for numerical columns in the dataset.
- Steps:
	- Load the dataset into a data analysis tool or programming environment (e.g., Python with pandas library).
	- Identify numerical columns in the dataset.
	- Calculate the mean, median, mode, and standard deviation for these columns.
	- Provide a brief interpretation of these statistics.

```python
# Importing Necesssary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset
sales_data = pd.read_csv('C:/DataScinceExlr/Assignments/3. Basic Stat - I/sales_data_with_discounts.csv')
```

```python
# Identify number columns
num_columns = sales_data.select_dtypes(include = [np.number]).columns
print('Numerical Columns: ',num_columns)
```

```python
# Calculating the mean, median and standard deviation,
# Since we are considering only numeric columns we will not calculate mode.

mean_values = sales_data[num_columns].mean()
median_values = sales_data[num_columns].median()
sd_values = sales_data[num_columns].std()

print('----------Descriptive Analytics for Numerical Columns-----------')
print('\n Mean \n', mean_values)
print('\n Median \n', median_values)
print('\n Standard Deviation \n', sd_values)
```

 #### Interpretation of these statistics:

1. Volume              
- Mean    = 5.066667 → On average, volume is  around 5 
- Median  =  4.000000  → The middle value is 4, which is less than the mean. This suggests the data might be right-skewed (a few higher values pulling the mean upward).
- Std Dev = 4.231602 → The values vary quite a bit around the mean (a spread of about 4.23 units), showing high variability relative to the average.   

2. Avg Price 
- Mean    =  10453.433333 --> On average, the price is quite high, around 10.4k.

3. Total Sales Value    
- Mean    = 33812.835556  --> On average, total sales are around 33.8k.
- Median  =  5700.000000  --> The middle value is much lower than the mean, indicating the distribution is strongly right-skewed (a few very large sales pushing the average upward).
- Std Dev = 50535.074173  --> Extremely high variability; sales values differ widely across transactions.

4. Discount Rate (%)
- Mean    =  15.155242 --> On average, the discount rate is about 15%
- Median  =  16.577766 --> The middle discount rate is slightly higher than the mean, suggesting the distribution may be left-skewed (more smaller values).
- Std Dev =  4.231602  --> The spread is about 4 percentage points, indicating moderate variation in discount rates.

5. Discount Amount

- Std Dev =  4509.902963 --> High variability, discounts vary widely across transactions.

6. Net Sales Value
- Mean    =  30466.336131 --> On average, the net sales value is around 30.4k.
- Median  =   4677.788059 --> The middle value is much lower than the mean, suggesting a strong right-skew (a few very high sales dominating).

### Data Visualization
- Objective: To visualize the distribution and relationship of numerical and categorical variables in the dataset

#### Histograms:
- Plot histograms for each numerical column.
- Analyze the distribution (e.g., skewness, presence of outliers) and provide inferences.


```python
# Importing necessary modules for data visualization
import matplotlib.pyplot as plt
```

```python
#Numeric column : Volume 
plt.figure(figsize=(6, 3))
plt.hist(sales_data['Volume'], bins=10, color='skyblue', edgecolor='black', alpha=0.7)

# Labels & title
plt.xlabel("Volume")
plt.ylabel("Frequency")
plt.title("Histogram of Volume")

# Mean & Median lines
mean_val = sales_data['Volume'].mean()
median_val = sales_data['Volume'].median()
plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=2, label=f'Mean = {mean_val:.2f}')
plt.axvline(median_val, color='green', linestyle='dotted', linewidth=2, label=f'Median = {median_val:.2f}')

# Grid + legend
plt.grid(axis='y', linestyle='--', alpha=0.7)  #To increase the readability
plt.legend()                                   #To explain lines&color

plt.tight_layout()    #automatically adjusts the spacing between plot elements
plt.show()
```

```python
plt.figure(figsize=(6, 3))
plt.hist(sales_data['Avg Price'], color='#609AA1', edgecolor='black', alpha = 0.5)

# Labels and Titles
plt.xlabel('Average Price')
plt.ylabel('Frequency')
plt.title('Avg Price')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 
plt.figure(figsize=(6,3))
plt.hist(sales_data['Total Sales Value'], color='#BFB4BF', edgecolor='black', alpha = 0.7)
plt.title('Total Sales Value')
plt.xlabel('Sales Value')
plt.grid(axis ='y', linestyle = '--', alpha = 0.7 )
plt.show()
```

```python
plt.figure(figsize=(6, 3))
plt.hist(sales_data['Discount Rate (%)'], color='#C9ED9D', edgecolor='black', alpha = 0.7)
plt.show()
```

### Box plot

```python
plt.figure(figsize=(15, 8))
# Create boxplots
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=sales_data[col], color="skyblue")
    plt.title(f"Boxplot of {col}")

plt.tight_layout()
plt.show()
```
