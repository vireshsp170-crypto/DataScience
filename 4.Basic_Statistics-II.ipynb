```python
patients.head()
```

```python
# 2. Selecting only relevant data for billing. [ 'PatientID', 'Department', 'Doctor', 'BillAmount' ]
patients = patients[['PatientID','Department', 'Doctor', 'BillAmount']]
```

```python
patients.head()
```

#### In the above cell, we can see that we have only the relevant data which is necessary for the billing.

#### We can see that there are two NaN in the BillAmount Column, We should replace that with the mean of the column since it is a numerical column

```python
# Calculate mean bill

mean_bill = patients['BillAmount'].mean()
print(mean_bill)
```

```python
# Replacing the NaN of BillAmount column entries with the mean value.
patients['BillAmount'] = patients['BillAmount'].fillna(mean_bill)
```

```python
patients.head()
```

#### We can see that in the above cell, we have successfully dropped the columns which are unnecessary for our data interpretation.

```python
# 4. Total bill amount per department
dept_bill_summary = patients.groupby('Department')['BillAmount'].sum()
print(dept_bill_summary)
```

```python
# 5. Remove duplicate patients (keep first entry)
patients.drop_duplicates(subset='PatientID', keep='first', inplace=True)
```

```python
patients.head()
```

```python
billing_data = pd.read_csv('C:/Users/Admin/Datasets/Billing_Data.csv')
```

```python
billing_data.head()
```

#### We can see that, in the above data there are no any duplicates.

```python
# Merge billing + patient info on PatientID

merged_data = pd.merge(patients,billing_data,  on='PatientID', how='outer')
```

```python
merged_data.head()
```

```python
merged_data.columns
```

```python
new_patient = pd.DataFrame([
    [ 107 , "Orthopedics" , "Dr. Kiran" , 10000.0 , 3500 , 6500],
    [ 108 , "Neurology",    "Dr. John",    6500.0,  1500 , 5000]
], columns=merged_data.columns)
```

```python
new_patient
```

```python
patients = pd.concat([merged_data, new_patient], ignore_index=True )
```

```python
patients
```

```python
# Final cleaned dataset with accurate billing info.

patients["FinalAmount"] = patients["BillAmount"] - patients["InsuranceCovered"]
print(patients)
```

### Ability to perform analytics



```python
dept_revenue = patients.groupby('Department')['FinalAmount'].sum()
print(dept_revenue)
```


### 1. Department level insights:
- Cardiology: 3000 + 3200 = 6200c
- Neurology: 4425 + 5000 = 9425
- Orthopedics: 5000 + 6500 = 11,500
- Dermatology: 4925
##### Orthopedics contributes the highest revenue.
##### Dermatology has the least total revenue (only one patient so far).

```python
dr_perf = patients.groupby('Doctor')['FinalAmount'].sum()
print(dr_perf)
```

### Plotting the data
