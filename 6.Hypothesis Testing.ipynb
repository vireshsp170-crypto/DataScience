## Estimation and Confidence Interval

```python
import statistics as st
import numpy as np
from scipy.stats import norm
```

```python
# sample size
n = 15
```

```python
# Given Data
sample_data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29] #in millions
```

### a. Build 99% Confidence Interval Using Sample Standard Deviation

```python
"""
In the given data, we are not aware of the population size.
So, when the population standard deviation is unknown and 'n' is too small, we will use t-distribution.

When is unknown and n is small, the sampling distribution of the sample mean has more spread than a normal distribution; 
the t-distribution (with df=n−1) accounts for that extra uncertainty.

"""
```

```python
# Degrees of freedom
df = n-1
df
```

```python
"""
Two-sided 99% CI critical t-value: 
t(0.995, 14)

(This is approximately 2.977)
"""
```

```python
t = 2.997
t
```

```python
# sample mean
s_mean = st.mean(sample_data)
s_mean
```

```python
s = st.stdev(sample_data)
s
```

```python
# Standard error 
SE =  s / np.sqrt(n)
SE
```

```python
#Margin of Error 

ME = t * SE
ME
```

```python
# The range or interval of confidence is : +or- ME 
R_E1 = s_mean - ME
R_E1
```

```python
R_E2 = s_mean + ME
R_E2
```

```python
"""
By above calculation we can see that, the range for Confidence interval is:
1.090 to 1.387
"""
```

#### “With 99% confidence, we estimate that the true mean durability falls between approximately 1.090 and 1.387 million characters, based on our sample and assumptions"

## Hypothesis Testing 

```python
# Sample mean of 25 restaurants
smean = 3050  

# Population mean (or hypothesized mean units per week)
x_mean = 600  

# Sample size
n = 25  

# Significance level
alpha = 0.05  
```

```python
# Theoretical mean weekly operating cost
theoretical_mean = 1000 + 5 * x_mean
```

```python
# Since cost increases by 5 per unit and x has std dev of 25 units
sigma = 5 * 25  # = 125
```

```python
# Null hypothesis (H0): mu = theoretical_mean
# Alternative hypothesis (H1): mu > theoretical_mean
# This represents the right-tailed z-test.
```

```python
# calculate the z-test
z_score = (smean - theoretical_mean) / (sigma / np.sqrt(n))
z_score
```

```python
# calculate the P-value 
p_value = 1 - norm.cdf(z_score)

print(f"Z-score: {z_score:.3f}")
print(f"P-value: {p_value:.3f}")
```

```python
if p_value < alpha:
    print("Reject the null hypothesis. The results indicate that weekly operating costs are significantly higher than predicted by the model.")
else:
    print("Fail to reject the null hypothesis. The results do not provide sufficient evidence that weekly operating costs are higher than predicted by the model.")
```
