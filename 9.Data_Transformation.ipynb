#### 1. Handling Missing Values

```python
data.isna().sum()
```

```python
# Check unique values in the 'occupation' column
data['occupation'].unique()
```

- The value '?' represents a missing or unknown value.

 ##### It means:
The person’s occupation was not recorded
Or the data was not available at the time of collection

```python
# Check unique values in each categorical column - Detect invalid placeholders like ' ?'
# This helps identify placeholders like '?', 'NA', 'NULL', etc.
for col in data.select_dtypes(include="object").columns:
    print(col)
    print(data[col].unique())
    print("-" * 40)
```

#### The unique values of categorical columns were analyzed to identify valid categories and detect missing placeholders such as '?'. Columns containing placeholders require cleaning before encoding and model training. 
#### namely:  'workclass', 'occupation' and 'native_country'.

```python
# Select categorical columns
cat_cols = data.select_dtypes(include='object').columns

# Remove leading/trailing spaces
for col in cat_cols:
    data[col] = data[col].str.strip()
```

```python
# Replace '?' with NaN
data.replace('?', np.nan, inplace=True)
```

```python
# Verify 
for col in data.select_dtypes(include="object").columns:
    print(col)
    print(data[col].unique())
    print("-" * 40)
```

```python
# Check missing (null) values in each column
data.isna().sum()
```

```python
# Check numeric columns for zero value(Check Zero Counts)
num_features = data.select_dtypes(include=['int64', 'float64']).columns

for feature in num_features:
    zero_count = (data[feature] == 0).sum()
    print(f"{feature} has {zero_count} zero values")    
```

```python
# Replace zero values with median for numeric columns
numeric_columns = data.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_columns:
    if (data[col] == 0).any():
        median_val = data.loc[data[col] != 0, col].median()
        data.loc[data[col] == 0, col] = median_val
```

```python
# Check categorical columns and replce  with median value
cat_col = data.select_dtypes(include="object").columns
```

```python
# Replace missing categorical values with mode
data[cat_col] = data[cat_col].apply(lambda x: x.fillna(x.mode()[0]))
```

```python
# Verify that there are no missing values in the dataset
data.isna().sum()
```

#### 1.Apply scaling techniques to numerical features

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# Apply standard scaling to numerical features
scaler = StandardScaler()
data_scaled = data.copy()
```

```python
# Apply standard scaling to numerical columns
num_col = data.select_dtypes(include=["int64", "float64"]).columns
data_scaled[num_col] = scaler.fit_transform(data_scaled[num_col])
```

```python
# Feature scaling ensures that all features contribute equally to the model
# Standard scaling helps achieve faster convergence during optimization
# It is commonly used when features follow (or are close to) a normal distribution
```

```python
# Applying Min-Max Scaling
mm_scaler = MinMaxScaler()
data_mm = data.copy()
```

```python
# Applying Min-Max scaling to numerical columns
# This scales values between 0 and 1 using min and max of each feature
data_mm[num_col] = mm_scaler.fit_transform(data[num_col])
```

```python
# Min-Max scaler is used when the data is NOT normally distributed
# It scales all values between 0 and 1
```

#### 2. Encoding Techniques: 

```python
# checking how many unique values in each categorical column
data[cat_col].nunique()
```

- If number of categories ≤ 5  → One-Hot Encoding
- If number of categories > 5  → Label Encoding


```python
# One-hot encoding for low-cardinality categorical columns (≤ 5 unique values)
low_cardinality_cols = [col for col in cat_col if data[col].nunique() <= 5]

data = pd.get_dummies(data, columns=low_cardinality_cols, drop_first=True)
```

```python
from sklearn.preprocessing import LabelEncoder

# High-cardinality categorical columns
high_cardinality_cols = [col for col in cat_col if col not in low_cardinality_cols]

label_encoder = LabelEncoder()

for col in high_cardinality_cols:
    data[col] = label_encoder.fit_transform(data[col])
```

#### 3. Feature Engineering: 

```python
# Create a new feature: Net Capital Gain
data["capital_net"] = data["capital_gain"] - data["capital_loss"]
```

```python
# Create a new feature: Work Intensity Score
data["work_intensity"] = data["hours_per_week"] * data["education_num"]
```

```python
# Compare skewness of all numerical features
num_cols = data.select_dtypes(include=["int64", "float64"]).columns
data[num_cols].skew().sort_values(ascending=False)
```

```python
# Capital gain shows extreme positive skewness with a long right tail
# Apply log transformation to reduce skewness
data["capital_gain_log"] = np.log1p(data["capital_gain"])
```
